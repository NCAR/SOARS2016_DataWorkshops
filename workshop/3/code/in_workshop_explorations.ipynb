```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline

# let's load the data in now
df = pd.read_csv("../data/sample_demo_weather_data_1981.csv")
```

```python
df[df['raintot']>=0].to_html("sample_demo_weather_data_1981_clean.html")
```

```python
df.head()
```

```python
df.groupby(['month']).describe()
```

```python
df_raintot = df[['raintot', 'tmaxavg','month']]
```

```python
df_raintot = df_raintot[df_raintot['raintot']>=0]
```

```python
plt1 = df_raintot.plot(secondary_y=['tmaxavg'])
```

```python
plt1.figure.savefig('aplot.png', dpi=600)
```

```python
plt2 = df_raintot.plot.scatter(x='month',y='raintot')
```

```python
plt2.figure.savefig("bplot.png", dpi=600)
```

```python
plt3 = df_raintot.boxplot(column='raintot', by='month')
```

```python
plt3.figure.savefig("cplot.png", dpi=600)
```
