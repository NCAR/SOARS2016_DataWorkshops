# Python File Basics

Opening a file in Python is very simple -- we simply need to write code like this to get the _File object_.


```python
f = open("../data/sample_data.txt")
```

Now we can actually work on the file.  At first, let's just read each line of the file and print it line by like like so:

```python
# iterate over file lines with for .. in ... : construct
for l in f:
    print l
```

While this is a legitimate way to open a file, the _recommended_ way is like this:

```python
# using the 'with' statement for opening files is the recommended way to access files
with open("../data/sample_data.txt") as f:
    for l in f:
        print l
```

## Reading a file as a string all at once

Reading a file line by line is useful in many scenarios, especially those where the file is large or that there is some additional processing that needs to happen as the file is being read, and possibly when the _entire_ file does not need to be read (e.g. you would like to stop processing the file once some condition is met).

But when the file is small enough you can load the file into a string all at once with `readlines()`:

```python
with open("../data/sample_data.txt") as f:
    file_string = f.readlines()
    
    # all lines as a list
    print file_string
    
    # the last line
    print file_string[-1]
    
    # the last 5 lines
    print file_string[-5:]
```

## Using the  `csv` library
If you want to save even more time processing files even further, use the `csv` library and you get to make the line splitting for free (and you can still yse any delimiter you would like

```python
import csv

with open("../data/sample_data.txt") as f:
    csv_reader = csv.reader(f, delimiter='\t')  # remember the default delimiter is ',', whereas the file we have is Tab ('\t')

    for l in csv_reader:
        print l
```

Now with `csv` we can look at each line and _column_ as if it were a list ... very nice.

## Taking things further using `DictReader` 

We can take our file reading one step further by using the `DictReader` which is again a convenience method that provides an even more streamlined way to read a CSV file.

Say the first line of the file is the header -- and in our case we've got `sample_data_with_header.txt` as an example.  In that file, the **first** line says `lat, lon` indicating latitude and longitude.  Now with `DictReader` such data is now accessible by the rows and columns more naturally.

Instead of using the index (0, 1, 2, ...) to indicate the column we want, we can call out the column by its name _as given by the header row_ (or first line of the file).  Let's see an example,

```python
with open('../data/sample_data_with_header.txt') as f:
    csv_file = csv.DictReader(f, delimiter='\t')
    
    for l in csv_file:
        print l['lat']
```
