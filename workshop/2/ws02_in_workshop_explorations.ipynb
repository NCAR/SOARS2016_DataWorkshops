## A diversion back to lists ...
In the last workshop we touched on lists, but what we didn't talk as much about, is how to actually slice up a list.  Lists are not unlike arrays in the way you might like to access the data in them.  Let's play ...

```python
lst = [1,2,3,4,5]
```

```python
lst
```

Now, lets say we wanted the second and third elements of the list (index 1 through 3, exclusive).  We would do something like this:

```python
lst[1:3]
```

Now just for the sake of the excercise, let's create a list of 100 numbers:

```python
lst = range(100)
```

The nice thing about the slicing in Python is that we can slice from the _end_ (negative numbers) as well.  This gives us the -5 though -3 items in the list of 100 numbers.

```python
lst[-5:-3]
```

Also, what's quite cool is that we can also slice by a _step_ or _interval_ using a third `:`.  So for example, if we wanted to see every other item in the list, we'd say something like 

    lst[::2] 

That is, show me *all* items in the list, skipping every other one.

Now we can also do the usual thing with lists and concatenate them.  Say we wanted to create the list of items where the first 50 were _every other item_ and then append the rest of the items from 51 onwards to the end.

Simple enough:

```python
lst[1:50:2] + lst[51:]
```

## Accessing individual rows in a Dataframe

Let's say for a second that we want to access an individual row in a Dataframe.  Intuitively, we might try something like this:

```python
df[1] # oops!
```

## Ooops! Maybe not!

Let's take a look at the error a second.  It looks like `KeyError` is telling us something.  In fact, what this means is that the Dataframe is expecting a **key** when using this form of the syntax.  Well -- what are the **keys**, you might ask ... they are actually the **columns**.  Thus, we used:

    df[_column_name_]
    
we might get something useful ...

```python
df['lat']
```

## OK, but that's not what I want ... how about that row???
Well going back to our _slicing_ syntax on lists, we might be compelled to try something like this:
    
    df[start_index:end_index]

```python
df[:1] # wow!
```

```python
df[0:10] # first 5 -- same as df.head(10)
```

Now we're ready to combine our __key-based__ lookup with our slices ...

```python
df['lat'][9:10]  # give me the 'lat' columns 10th value (remember 0 based indexing)
```

## Selected data YOUR way
Index-based slicing works fine when you know what you want -- that is, you know where the data lives, but that's usually not always the case (in fact, it is more often NOT the case, and definitely not the case when you have A LOT of data _from someone else_, where you aren't really familiar with it ... hence *Exploratory* Data Analysis).

Let's play for a minute.  Say you want to find out all data in your Datafram where the month was equal to 1.  Recall, that Python equivalence is performed with `==`.  Let' see what happens when we do this:

```python
df['month'] == 1
```

This asks -- what elements in Dataframe have a `month` *equal to* `1`.  What got returned was a `Series` object that included the index (row) and the result of the query we wanted for each item in the Dataframe.  The first few values we'll notice have `True`, while the remaining contain `False`.  Thus, we use this as our index to the Dataframe (`df`) we're asking it to return us the rows in our Dataframe (`df`) where the rows/index of our query are `True`.  We get:

```python
df[df['month']==1]
```

Awesome!  We now have all the rows in `df` that are only in the month of January!

A more sophisticated query asking for the first 10 days of May.  **Notice** the parenthesis aroung each filter separated by a `&` ... it must look this way or you will run in to problems.

```python
df[(df['month']==5) & (df['day']<=10)] # you can add more to this if you'd like -- you are not limited to the number if & separated filters
```

## Filtering "bad" data
Before we noted that the minimum rainfall for a few items in our data we're less than 0.  Rainfall less than 0!  Not so fast ... that sounds remarkable, but worse than that ... SO much lower than 0 that it skewed the minimum so low, **it too** went negative.  That makes no sense and we don't want that in our data.  So we're going to use our new filtering friend to ask ONLY for the cells that have raintotal at or more than 0:

    df['raintot'] > 0
    
Now then we'll save this in a new variable called `df_good`.


```python
df_good = df[df['raintot'] > 0]
```

## More filters with count()
Let's say we want to know a little more about the data in May ... and wanted to count the number of days where the `raintot` was more than the average for the year.

First, let's use `df_good` to get a good average:

```python
df_good_mean = df_good.describe()['raintot']['mean']
print df_good_mean
```

Good!  Now let's try filter for the days in May less than average raintot:

```python
df_good[(df_good['month']==5) & (df_good['raintot']<=df_good_mean)]
```

Now more than the average:

```python
df_good[(df_good['month']==5) & (df_good['raintot']>=df_good_mean)]
```

## Counting made easy with `count()`
Now that's cool but what if we wanted to know the percentage of days greater than and less than the average.

We can use `count()` to guide us to the answer.

```python
total_may_measurements = df_good[df_good['month']==5]['raintot'].count()
total_may_above_avg    = df_good[(df_good['month']==5) & (df_good['raintot']>=df_good_mean)]['raintot'].count()
total_may_below_avg    = df_good[(df_good['month']==5) & (df_good['raintot']<df_good_mean)]['raintot'].count()

print "May\n=====\n{} total rain measurements.\n{} days ({}%) above average.\n{} days ({}%) below average.\n" \
        .format(total_may_measurements, total_may_above_avg, 100.*total_may_above_avg/total_may_measurements,\
                total_may_below_avg, 100.*total_may_below_avg/total_may_measurements)
```

## Summing made easy with `sum()`
Much like `count()` we can use the function `sum()` to learn some things about the sum of an aggregate of data.

Say we want to know what the **total** rainfall was in May.  

Easy enough:

```python
df_good['raintot'][df_good['month']==5].sum()
```
