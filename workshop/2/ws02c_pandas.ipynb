## Pandas in 2 minutes
Pandas is a standard entry point into data manipulation in Python.  It provides a set of standard tools for loading, manipulating and geting basic statistical information about your data.  The documentation for Pandas is excellent [directly on their website](http://pandas.pydata.org) ... and the tutorials are highly recommended.

In essence Pandas provides you the ability to:

1. load data from various sources (CSV, databases, HTTP, etc.),
2. manipulate, reshape and recompose the loaded data (e.g. remove bad data, filter data),
3. do basic statistical analysis on the data,
4. store the data back down into a datasource (CSV, database, etc.).

You will find Pandas a bundle of joy once you get the hang of it!

## Dataframe data access

Accessing a dataframe is not unlike accessing a list -- the basic form of slicing the dataframe __by rows__ looks like this:

    df[start_index:end_index:skip_size]

Let's take a look:

```python
df[741:742] # slicing by rows outputing only row index 741 (row number 742)
```

```python
df[1:20:2]  # slicing by rows skipping every 2nd item from 1 to 20
```

### More complex slicing

Now what gets returned when we use this form is another Dataframe. Accessing the rows and columns (also known as _indices_) might also be done a different way if you need a more complex slice.  Say you wanted the first and last 20 items in your data frame.  We can simply use `pandas.concat(list_of_slices)`  to get the data.

```python
pd.concat([df[1:20], df[-20:]])  # could you do the same with head and tail? Try it!
```

### Slicing by filters

```python
df[df['month']==4].head() # only April .. and just the first 5 lines with head() by default
```

What if you want more  than one filter, say you want just the first 10 days of April:

```python
df[(df['month']==4) & (df['day']<=10)]
```

### Using describe data &#longdash; a short diversion

Say we wanted to plot the bounding box on a map for the lat/lon data in our data set.  One such (easy) way to do this is to look at the `min()` and `max()` values of the `lat` and `lon` like so:

```python
min_lon, max_lon, min_lat, max_lat = (df['lon'].min(), df['lon'].max(), df['lat'].min(), df['lat'].max())
```

Now we're going to have some fun ... more about this will be revealed in Workshop 4, but we're going to use a library called [Folium]((https://folium.readthedocs.io/en/latest/) ), which allows us to build interactive maps.  We will just show a quick example of how to build a bounding box, and invite you to explore more of the details of what Folium can do!

```python
import folium

lat = (min_lat+max_lat)/2
lon = (min_lon+max_lon)/2
zoom_start = 5

mapa = folium.Map(location=[lat, lon], zoom_start=zoom_start)

locations = [
    [(min_lat, min_lon),(min_lat, max_lon)],
    [(min_lat, max_lon),(max_lat, max_lon)],
    [(max_lat, max_lon),(max_lat, min_lon)],
    [(max_lat, min_lon),(min_lat, min_lon)]
]

c3 = folium.MultiPolyLine(locations=locations, color='gray', weight=2, opacity=.5)

for c in [c3]:
    mapa.add_children(c)
mapa
```
